```python
data['married'] = data['married'].map({'single':0, 'married':1})
data['car_ownership'] = data['car_ownership'].map({'no':0, 'yes':1})
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# the independent variables set
X = data[['married', 'car_ownership','experience','current_job_years','current_house_years','age','income']]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columnsp
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

```python
y=df['risk_flag'] 
#removed age,married and current_job_years
cols=['income','car_ownership','married'] #'current_job_years','age']
X=df[cols]
```

```python
oversample = SMOTE(sampling_strategy=0.5)
X, y = oversample.fit_resample(X, y)
```

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=48)

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=48)
```
